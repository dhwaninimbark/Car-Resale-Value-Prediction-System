```python
import pandas as pd

excel1='ahmedabad.csv'
excel2='mumbai.csv'
excel3='hyderabad.csv'
excel4='kolkatta.csv'
excel5='pune.csv'
excel6='delhi.csv'


df1=pd.read_csv(excel1)
df2=pd.read_csv(excel2)
df3=pd.read_csv(excel3)
df4=pd.read_csv(excel4)
df5=pd.read_csv(excel5)
df6=pd.read_csv(excel6)

df1['city'] = "Ahmedabad"
df2['city'] = "Mumbai"
df3['city'] = "Hyderabad"
df4['city'] = "Kolkata"
df5['city'] = "Pune"
df6['city'] = "Delhi"

values1=df1[['model','fuel','km_driven','resale_val','city']]
values2=df2[['model','fuel','km_driven','resale_val','city']]
values3=df3[['model','fuel','km_driven','resale_val','city']]
values4=df4[['model','fuel','km_driven','resale_val','city']]
values5=df5[['model','fuel','km_driven','resale_val','city']]
values6=df6[['model','fuel','km_driven','resale_val','city']]


dataframes=[values1,values2,values3,values4,values5,values6]

join=pd.concat(dataframes)


join.to_excel("md1.xlsx")
```
